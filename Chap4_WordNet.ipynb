```python
import nltk
import spacy_udpipe
spacy_udpipe.download("en")
nltk.download('wordnet')
nltk.download('sentiwordnet')
from nltk.corpus import wordnet
from nltk.corpus import sentiwordnet as swn

nlp = spacy_udpipe.load("en")
```

```python
word = 'joy'
synsets = wordnet.synsets(word)

print(f"The word '{word}' has {len(synsets)} synsets in WordNet.")
for synset in synsets:
    print(f"- {synset.name()}: {synset.definition()}")
```

```python
doc = nlp("This is a sample sentence.")
for token in doc:
  if not token.is_punct:
    print(token.lemma_, end="/")
```

```python
sent = "The girl grew up locked on a hidden tower, her mom, \
the witch, used to lie to her saying that outside the tower, \
there were bad people that wanted to hurt her."
doc = nlp(sent)
for token in doc:
  if not token.is_punct:
    print(token.lemma_, end="/")
```

```python
for token in doc:
  if not token.is_punct:
    synsets = list(swn.senti_synsets(token.lemma_))
    if synsets:
      sst = synsets[0]  # Take the first synset
      pos, neg, obj = sst.pos_score(), sst.neg_score(), sst.obj_score()
      print(f"Token: {token.lemma_} \tPos: {pos}\tNeg: {neg}\tObj: {obj}")
```

```python
pos_count, neg_count = 0, 0
for token in doc:
  if not token.is_punct:
    synsets = list(swn.senti_synsets(token.lemma_))
    if synsets:
      sst = synsets[0]  # Take the first synset
      pos, neg = sst.pos_score(), sst.neg_score()
      match (pos > neg, neg > pos):
         case (True, False):  pos_count += 1
         case (False, True):  neg_count += 1

print("正向字個數：",pos_count," 負向字個數：", neg_count)
```

```python
pos_count, neg_count, obj_count = 0, 0, 0
for token in doc:
  if not token.is_punct:
    synsets = list(swn.senti_synsets(token.lemma_))
    if synsets:
      sst = synsets[0]  # Take the first synset
      pos, neg, obj = sst.pos_score(), sst.neg_score(), sst.obj_score()
      match (pos > neg, pos > obj, neg > obj):
         case (True, True, _):  pos_count += 1
         case (_, False, True): neg_count += 1
         case _:                obj_count += 1
print("正向字個數：",pos_count,
      " 負向字個數：", neg_count, " 中性字個數：", obj_count)
```

```python
pos_value, neg_value, obj_value = 0, 0, 0
for token in doc:
  if not token.is_punct:
    synsets = list(swn.senti_synsets(token.lemma_))
    if synsets:
      sst = synsets[0]  # Take the first synset
      pos_value += sst.pos_score()
      neg_value += sst.neg_score()
      obj_value += sst.obj_score()

print("正向字數值：",pos_value," 負向字數值：", neg_value,
      " 中性字數值：", obj_value)
```

```python
def get_sentiment_scores(word):
    synsets = list(swn.senti_synsets(word))
    if not synsets:
        return None
    # Take the first synset (most common usage)
    synset = synsets[0]
    return synset.pos_score(), synset.neg_score(), synset.obj_score()

# Process each token
for token in doc:
    if not token.is_punct:
        sentiment_scores = get_sentiment_scores(token.lemma_)
        if sentiment_scores:
            pos, neg, obj = sentiment_scores
#            print(f"Token: {token.lemma_} \tPos: {pos}, Neg: {neg}, Obj: {obj}")
        else:
            print(f"Token: {token.lemma_} \tNo sentiment scores found")
```
